## First Steps

#### Load `FASTA` file 

```python
from Bio import SeqIO
```

```python
records = list(SeqIO.parse("data/creb.fasta", "fasta"))
dna = records[0]
print(dna.name)
print(dna.description)
print(dna.seq[:100])
```

```python
print(f'Length of this DNA sequence is {len(dna)}')
```

#### Reverse Complement

```python
from Bio.Seq import Seq
```

```python
creb_dna = Seq(dna.seq)
print(creb_dna.reverse_complement())
```

#### Calculate GC content

```python
def calculate_content(string, nucleotide):
    processed=string.lower()
    processed_nucl = nucleotide.lower()
    print(processed_nucl)
    if processed_nucl !='a' or processed_nucl !='t' or processed_nucl !='g' or processed_nucl !='c':
        print('Enter a viable nucleotide Hint:A,T,G,C')
    return processed.count(processed_nucl)/len(string)
```

```python
g_content = calculate_content(dna.seq[:],'G')
c_content = calculate_content(dna.seq[:],'C')
print(f'GC content of sequence is {(g_content+c_content)*100}')
```
